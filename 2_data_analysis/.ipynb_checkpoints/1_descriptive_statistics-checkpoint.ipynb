```python
import pandas as pd
from plotnine import *
```

```python
# Load review datasets
health_df = pd.read_csv("../0_data/healthcare_reviews.csv")
infra_df = pd.read_csv("../0_data/infrastructure_reviews.csv")
```

## Overall Rating

```python
health_df["source"] = "health_df"
infra_df["source"] = "infra_df"

df_combined = pd.concat([health_df, infra_df], ignore_index=True)
# Remove NAs
df_combined["rating"] = pd.to_numeric(df_combined["rating"], errors="coerce")
df_clean = df_combined.dropna(subset=["rating"])

rating = (ggplot(df_clean, aes(x="rating", fill="source"))
    + geom_histogram(alpha=0.5, bins=20, position="identity")
    + labs(
        title="Histogram of Ratings for Healthcare/Infrastructure",
        x="Rating",
        y="Count"
    )
    + theme_bw())

rating.show()
```

```python
rating.save("../3_outputs/1_rating_histogram.png", width=8, height=5, dpi=600)
```

## Rating by dimension

```python
# Select and reshape dataframes
cols = ["career_opportunities", "compensation_and_benefits", "senior_management",
        "work_life_balance", "culture_and_values", "diversity_and_inclusion"]

health_long = (health_df[cols].assign(source="Healthcare")
               .melt(id_vars="source", var_name="dimension", value_name="rating"))

infra_long = (infra_df[cols].assign(source="Infrastructure")
              .melt(id_vars="source", var_name="dimension", value_name="rating")
)

combined_long = pd.concat([health_long, infra_long], ignore_index=True)

# Drop NAs
combined_long["rating"] = pd.to_numeric(combined_long["rating"], errors="coerce")
combined_long = combined_long.dropna(subset=["rating"])
combined_long = combined_long[(combined_long["rating"] >= 1) & (combined_long["rating"] <= 5)]
```

```python
# Count how often each rating appears per dimension
# Use percent instead of count because the review numbers are not the same for the two industries
count_df = (
    combined_long
    .groupby(["source", "dimension", "rating"])
    .size()
    .reset_index(name="count")
)
```

```python
# Mapping dimension names for better visualization
label_map = {
    "career_opportunities": "Career Opportunities",
    "compensation_and_benefits": "Compensation & Benefits",
    "culture_and_values": "Culture & Values",
    "diversity_and_inclusion": "Diversity & Inclusion",
    "senior_management": "Senior Management",
    "work_life_balance": "Work-Life Balance"
}

count_df["dimension"] = count_df["dimension"].map(label_map)
```

```python
# Create a line graph
rating_by_dimension = (
    ggplot(count_df, aes(x="rating", y="percent",
                         color="dimension", linetype="source"))
    + geom_line(size=1)
    + geom_point(size=2)
    + scale_x_continuous(breaks=[1,2,3,4,5])
    + facet_wrap("~dimension", scales="free_y", ncol=2)
    + labs(
        title="Counts of Ratings by Dimension",
        x="Rating (15)",
        y="Count"
    )
    + theme_bw()
)


rating_by_dimension.show()
```

```python
rating_by_dimension.save("../3_outputs/2_rating_by_dimension.png", dpi=600)
```

## Attitude towards companies

```python
# Prepare the long dataframe
sent_cols = ["recommend", "ceo_approval", "business_outlook"]

health_sent = (
    health_df[sent_cols]
    .assign(source="Healthcare")
    .melt(id_vars="source", var_name="variable", value_name="value")
)

infra_sent = (
    infra_df[sent_cols]
    .assign(source="Infrastructure")
    .melt(id_vars="source", var_name="variable", value_name="value")
)

sent_long = pd.concat([health_sent, infra_sent], ignore_index=True)

# Clean and ensure numeric
sent_long["value"] = pd.to_numeric(sent_long["value"], errors="coerce")
sent_long = sent_long.dropna(subset=["value"])

# Keep valid values only (-1, 0, 1)
sent_long = sent_long[sent_long["value"].isin([-1, 0, 1])]

# Count and convert to percentage
count_df = (
    sent_long
    .groupby(["source", "variable", "value"])
    .size()
    .reset_index(name="count")
)

# Total counts per source  variable
count_df["total"] = (
    count_df.groupby(["source", "variable"])["count"]
    .transform("sum")
)

# Compute percentages
count_df["percent"] = (count_df["count"] / count_df["total"]) * 100

count_df['variable'] = pd.Categorical(
    count_df['variable'],
    categories=["recommend", "ceo_approval", "business_outlook"],
    ordered=True
)

var_labels = {
    "recommend": "Recommend",
    "ceo_approval": "CEO Approval",
    "business_outlook": "Business Outlook"
}

company_attitude = (
    ggplot(count_df, aes(x="value", y="percent", fill="source"))
    + geom_col(position="dodge")
    + facet_wrap("~variable", labeller=lambda x: var_labels.get(x, x))
    + scale_x_continuous(breaks=[-1, 0, 1], labels=["Negative", "Neutral", "Positive"])
    + labs(
        title="Sentiment Comparison Across Industries",
        x="Sentiment Category",
        y="Percentage (%)"
    )
    + theme_bw()
)

company_attitude.show()
```

```python
company_attitude.save("../3_outputs/3_company_attitude.png", dpi=600)
```

```python
# The graph is not so intuitive at first glance, let's make some modifications
# Hightlight 0 for healthcare, -1 and 1 for infrastructure
def assign_color(row):
    if row["source"] == "Healthcare":
        return "Highlight" if row["value"] == 0 else "Shallow"
    else:  # Infrastructure
        return "Highlight" if row["value"] in [-1, 1] else "Shallow"

count_df["highlight"] = count_df.apply(assign_color, axis=1)

company_attitude_highlighted = (
    ggplot(count_df, aes(x="value", y="percent", fill="source", alpha="highlight"))
    + geom_col(position="dodge")
    + facet_wrap("~variable", labeller=lambda x: var_labels.get(x, x))
    + scale_x_continuous(
        breaks=[-1, 0, 1],
        labels=["Negative", "Neutral", "Positive"]
    )
    + scale_alpha_manual(values={"Highlight": 1.0, "Shallow": 0.45})
    + labs(
        title="Attitudes Towards Companies, Industry Comparison",
        x="Sentiment Category",
        y="Percentage (%)",
        alpha="Intensity"
    )
    + theme_bw()
)

company_attitude_highlighted.show()
```

```python
company_attitude_highlighted.save("../3_outputs/4_company_attitude_highlighted.png", dpi=600)
```
